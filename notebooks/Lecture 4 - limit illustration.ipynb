```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-10,10,20000)
dx = x[1]-x[0]
psi1 = np.exp(-x**2/2)
psi2 = psi1 + np.exp(-(x+5)**2*5)

def distances(psi1,psi2):
    delta = psi1-psi2
    L2 = np.trapz(delta**2,x)**0.5
    L1 = np.trapz(np.abs(delta),x)
    Linf = np.max(np.abs(delta))
    return {'L1': L1, 'L2': L2, 'Linf': Linf}

print( distances(psi1,psi2) )

plt.figure()
plt.plot(x,psi1)
plt.plot(x,psi2)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

# Define the function f(x, y) = (x^2 * y) / (x^2 + y^2)
def f(x, y):
    return (x**2 * y) / (x**2 + y**2) * np.exp(-x**2 - y**2)

# Create a grid of points
x = np.linspace(-4, 4, 400)
y = np.linspace(-4, 4, 400)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Set up the figure and 3D axis
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
surf = ax.plot_surface(X, Y, Z, cmap='plasma', edgecolor='none', shade=True)

# Add a color bar for the surface
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=5)

# Set axis labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z = f(x, y)')

# Set title
ax.set_title(r'Surface plot of $f(x, y) = \exp(-x^2-y^2)\frac{x^2 y}{x^2 + y^2}$')

plt.savefig('3dplot.png')
# Show the plot
plt.show()
```
