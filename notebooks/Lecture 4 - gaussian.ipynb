```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
%matplotlib notebook
```

```python
def f(x,y):
    #return (x**3 - y**2*x - 2*x*y**2)/(x**2 + y**2)
    return np.exp(-(x**2 + y**2))
    
```

```python
def plot_fun_2d(f, x_lim, y_lim, nx=200,ny=200, frange = (-.75,.75)):
    
    x = np.linspace(x_lim[0], x_lim[1], nx)
    y = np.linspace(y_lim[0], y_lim[1], ny)
    xx,yy = np.meshgrid(x,y)

    z = f(xx,yy)
    print (z.shape)
    
    fig = plt.figure()
#    ax = fig.add_subplot(111, projection='3d')
#    ax.plot_surface(xx,yy,z)
    plt.imshow(z,extent=[*x_lim, *y_lim])
    plt.colorbar()
    CS = plt.contour(z,levels=np.linspace(*frange,21), colors='w',extent = [*x_lim, *y_lim])
    plt.gca().clabel(CS, inline=True, fontsize=10)
    plt.xlabel('x')
    plt.ylabel('y')

```

```python
x_lim = (-2,2)
y_lim = (-2,2)
plot_fun_2d(f, x_lim, y_lim)
```

```python
plt.savefig('gaussian-2d.pdf')
```

```python
def surf_plot(f,x_lim=(-1,1),y_lim=(-1,1)):
    nx = 200
    ny = 200
    x = np.linspace(x_lim[0], x_lim[1], nx)
    y = np.linspace(y_lim[0], y_lim[1], ny)
    xx,yy = np.meshgrid(x,y)

    z = f(xx,yy)
    
    fig = plt.figure(figsize=(4,3))
    ax = fig.add_subplot(111, projection='3d')
    ax.plot_surface(xx,yy,z)
#     plt.imshow(z,extent=[*x_lim, *y_lim])
#     plt.colorbar()
#     CS = plt.contour(z,levels=np.linspace(*frange,21), colors='w',extent = [*x_lim, *y_lim])
#     plt.gca().clabel(CS, inline=True, fontsize=10)
    plt.xlabel('x')
    plt.ylabel('y')

```

```python
surf_plot(lambda x,y: x**2 + y**2)
plt.savefig('minimum.pdf')
surf_plot(lambda x,y: x**2 - y**2)
plt.savefig('saddle.pdf')
surf_plot(lambda x,y: -x**2 - y**2)
plt.savefig('maximum.pdf')
```

```python
surf_plot(lambda x,y: ((1-x -1j*y)**(-1)).real, x_lim=(0,1.1), y_lim=(0,1.1))
surf_plot(lambda x,y: ((1-x -1j*y)**(-1)).imag)
```
