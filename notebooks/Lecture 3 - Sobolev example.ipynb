```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
%matplotlib inline

# Set up matplotlib
%matplotlib inline
# set default figure size
plt.rcParams['figure.figsize'] = [6, 4]
# set tick label size 
plt.rcParams.update({'font.size': 10})
# set x any y label size and title size
plt.rcParams.update({'axes.labelsize': 15})
plt.rcParams.update({'axes.titlesize': 15})
# add grid by default
plt.rcParams['axes.grid'] = True
# set legend size to 10
plt.rcParams.update({'legend.fontsize': 10})
```

```python
x = np.linspace(-10,10,50000)

def f(x,n):
    return np.exp(-x**2/2)* (1 + np.sin(n*x)/n**.5)

def df(x,n):
    return -x * np.exp(-x**2/2) * (1 + np.sin(n*x)/n**.5) + np.cos(n*x) * np.exp(-x**2/2) * n**.5
```

```python
plt.figure()
plt.plot(x, f(x,30), label='n=30')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) for n=30')
plt.show()
```

```python
plt.figure()
plt.plot(x, f(x,1000), label='n=1000')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) for n=1000')
plt.show()
```

```python
plt.figure()
plt.plot(x, df(x,1000), label='n=1000')
plt.plot(x, df(x,30), label='n=30')
plt.legend()
plt.title('df(x)/dx for n=30 and n=1000')
plt.xlabel('x')
plt.show()
```

```python
def L2_norm(n):
    return np.sqrt(np.trapz(f(x,n)**2, x))

def W12_norm(n):
    return L2_norm(n) + np.sqrt(np.trapz((df(x,n)**2), x))

```

```python
n_list = np.arange(1,1000)
L2_norms = [L2_norm(n) for n in n_list]
W12_norms = [W12_norm(n) for n in n_list]

print(L2_norms)

plt.figure()
plt.plot(n_list, L2_norms, label='L2 norm')
plt.plot(n_list, W12_norms, label='W12 norm')
plt.xlabel('n')
plt.title('L2 norm and W12 norm')
plt.legend()
plt.show()
```
