```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
%matplotlib notebook
```

```python
def f(x,y):
    #return (x**3 - y**2*x - 2*x*y**2)/(x**2 + y**2)
    return (x**2 - y**2) / (x**2 + y**2)**0.5 * np.exp(-0.5*(x*x+y*y))
    
```

```python
def plot_fun_2d(f, x_lim, y_lim, nx=200,ny=200, frange = (-.75,.75)):
    
    x = np.linspace(x_lim[0], x_lim[1], nx)
    y = np.linspace(y_lim[0], y_lim[1], ny)
    xx,yy = np.meshgrid(x,y)

    z = f(xx,yy)
    print (z.shape)
    
    fig = plt.figure()
#    ax = fig.add_subplot(111, projection='3d')
#    ax.plot_surface(xx,yy,z)
    plt.imshow(z,extent=[*x_lim, *y_lim])
    plt.colorbar()
    CS = plt.contour(z,levels=np.linspace(*frange,21), colors='w',extent = [*x_lim, *y_lim])
    plt.gca().clabel(CS, inline=True, fontsize=10)
    plt.xlabel('x')
    plt.ylabel('y')

```

```python
x_lim = (-2,2)
y_lim = (-2,2)
plot_fun_2d(f, x_lim, y_lim)
```

```python
plt.savefig('function_2d.pdf')
```

```python
def f(x,y):
    #return (x**3 - y**2*x - 2*x*y**2)/(x**2 + y**2)
    return np.sign(x)*np.sign(y)*(np.abs(x)*np.abs(y))**(1/3)
```

```python
x_lim = (-2,2)
y_lim = (-2,2)
plot_fun_2d(f, x_lim, y_lim, frange = (-1.5,1.5))
```

```python
plt.savefig('pdiff-example1.pdf')
```
