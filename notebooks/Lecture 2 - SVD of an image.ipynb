```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
%matplotlib inline
from imageio.v2 import imread 
```

```python
A = imread('parrot.jpg').astype(float)[::2,::2,0]
print(f'Shape of P: {A.shape}')
plt.imshow(A,cmap='gray')
```

```python
U, S, V = np.linalg.svd(A)
```

```python
print(f'Shape of U: {U.shape}, length of S: {len(S)}, shape of V: {V.shape}')
```

```python
plt.figure()
plt.semilogy(S, '.-')
plt.xlabel('i')
plt.ylabel('sigma_i')
plt.title('Singular values of matrix')
```

```python
def approx_image(nmax):
    """Approximate image using nmax singular values"""
    compression_ratio = (U[:,:nmax].size + S[:nmax].size + V[:nmax,:].size) / A.size
    return U[:,:nmax] @ np.diag(S[:nmax]) @ V[:nmax,:], compression_ratio
```

```python
plt.figure()
A_approx, compression_ratio = approx_image(50)
plt.imshow(A_approx,cmap='gray')
plt.title(f'Approximation with 10 singular values, compression ratio: {compression_ratio:.3f}') 
```
