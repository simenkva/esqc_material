```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
%matplotlib inline

# Set up matplotlib
%matplotlib inline
# set default figure size
plt.rcParams['figure.figsize'] = [6, 4]
# set tick label size 
plt.rcParams.update({'font.size': 10})
# set x any y label size and title size
plt.rcParams.update({'axes.labelsize': 15})
plt.rcParams.update({'axes.titlesize': 15})
# add grid by default
plt.rcParams['axes.grid'] = True
# set legend size to 10
plt.rcParams.update({'legend.fontsize': 10})
```

```python
x = np.linspace(-10,10,50000)

def f(x,n):
    return np.exp(-x**2 * n )
```

```python
plt.figure()
plt.plot(x, f(x,0.5), label='n=0.5')
plt.plot(x, f(x,10), label='n=10')
plt.plot(x, f(x,1000), label='n=1000')

plt.xlabel('x')
plt.title('f_n(x)')
plt.legend()
plt.show()
```

```python
def L1_norm(n):
    return np.trapz(np.abs(f(x,n)), x)


def Linf_norm(n):
    return np.max(np.abs(f(x,n)))


```

```python
n_list = np.linspace(1,100, 20)**2
L1_norms = [L1_norm(n) for n in n_list]
Linf_norms = [Linf_norm(n) for n in n_list]



plt.figure()
plt.plot(n_list, L1_norms, '*-', label='L1 norm')
plt.plot(n_list, Linf_norms, '*-', label='Linf norm')
plt.xlabel('n')
plt.ylabel('Norm')
plt.title('Norms of f_n(x)')
plt.legend()
plt.show()
```
