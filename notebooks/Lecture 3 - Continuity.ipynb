```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  
%matplotlib notebook
```

```python
x0 = np.linspace(-2,2,500)
xx,yy = np.meshgrid(x0,x0)
z = xx**2 / (xx**2 + yy**2)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(xx,yy,z)
```
